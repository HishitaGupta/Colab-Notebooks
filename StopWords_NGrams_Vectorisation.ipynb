# N-GRAMS

```python
from nltk import ngrams

nltk.download('punkt')
```

```python
def generate_ngrams(text, n):
  tokens = word_tokenize(text)
  n_grams = list(ngrams(tokens,n))
  return n_grams
```

```python
txt ="usually happens when installing spaCy because one of its dependencies (blis, thinc, or pydantic) fails to compile locally. Would you like me to show how to install a specific spaCy model (like en_core_web_sm) after fixing this?"

output = generate_ngrams(txt,2)
```

# TF-IDF - Word Embedding Technique

- Ecommerce Data
- 4 labels : Household, Electronics, Clothing & Books
- Task is to create a classification model that can predict a given descriptionof a product and classify theem as one of the labels using tfidf vectorisation technique.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Machine learning is a subset of artificial intelligence.",
    "Deep learning allows computers to learn complex patterns from data.",
    "Natural language processing helps computers understand human language.",
    "Artificial intelligence and data science are transforming industries.",
    "Supervised learning uses labeled datasets to train algorithms.",
    "Unsupervised learning finds hidden structures in unlabeled data.",
    "Reinforcement learning trains models using rewards and penalties.",
    "Neural networks are at the core of deep learning techniques.",
    "Data preprocessing is essential for building accurate models.",
    "AI applications are growing rapidly across various domains."
]

```

```python
v = TfidfVectorizer()
v.fit(corpus)
transform_output = v.transform(corpus)
```

```python
v.vocabulary_
```

```python
# print the idf of each word

all_feature_names = v.get_feature_names_out()

for word in all_feature_names:
  indx = v.vocabulary_.get(word)

  # get the score
  idf_score = v.idf_[indx]

  print(f"{word}:{idf_score}")
```

```python
# print the transformed out from tf-idf
print(transform_output.toarray())
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.label.value_counts()
```

```python
df['label_num'] = df['label'].map({
    'Household':0,
    'Electronics':1,
    'Clothing & Accessories':2,
    'Books':3
})
```

```python
df.head()
```

```python
# Train Test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Text,df.label_num,test_size =0.2 )
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# Tf idf Vectorizer

tf=  TfidfVectorizer()
X_train_tf = tf.fit_transform(X_train)
X_test_tf = tf.transform(X_test)
```

```python
# Classification Model

from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train_tf,y_train)

y_pred = clf.predict(X_test_tf)
```

```python
X_test_tf.shape
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```
