```python
import re

def expand_contractions(text):
  contractions_pattern ={
      r"(?i)can't":"cannot",
      r"(?i)won't" :"willnot",
      r"(?i)couldn't":"couldnot",
      r"(?i)shouldn't":"shouldnot",
      r"(?i)it's":"it is",
      r"(?i)were'nt":"were not",
      }
  for contraction, expansion in contractions_pattern.items():
      text = re.sub(contraction,expansion,text)
      return text

txt = "I can't believe it's already raining."
new_txt = expand_contractions(txt)

print(new_txt)
```
